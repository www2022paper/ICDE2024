## limitation1: covariate shift

```python
# output pic

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

DRP_avg_aucc_cost_curve = pd.read_csv("../figure/DRP_avg_aucc_cost_curve_1.0_shift.csv")
rDRP_avg_aucc_cost_curve = pd.read_csv("../figure/DRP_avg_aucc_cost_curve_1.0.csv")

DRP_aucc = np.sum(DRP_avg_aucc_cost_curve['delta_reward'].values) / (DRP_avg_aucc_cost_curve['delta_reward'].values[-1] * 201)
rDRP_aucc = np.sum(rDRP_avg_aucc_cost_curve['delta_reward'].values) / (rDRP_avg_aucc_cost_curve['delta_reward'].values[-1] * 201)

# output data results
print(f"DRP aucc: {DRP_aucc:.4f}")
print(f"rDRP aucc: {rDRP_aucc:.4f}")
improve = 100*(rDRP_aucc - DRP_aucc)/DRP_aucc
print(f"{improve:.2f}%")

np.sum(rDRP_avg_aucc_cost_curve['delta_reward'].values) / (rDRP_avg_aucc_cost_curve['delta_reward'].values[-1] * 201)

x_index = rDRP_avg_aucc_cost_curve['delta_cost'].values
y_random = rDRP_avg_aucc_cost_curve['delta_reward'].values[-1] / x_index[-1] * x_index

x_normalization = x_index[-1]
y_normalization = rDRP_avg_aucc_cost_curve['delta_reward'].values[-1]

y_normalization_DRP = DRP_avg_aucc_cost_curve['delta_reward'].values[-1]


# random
plt.plot(x_index / x_normalization, y_random / y_normalization, color='#000000', label='Random', linewidth=1.5, linestyle='--')

# DRP
plt.plot(x_index / x_normalization, DRP_avg_aucc_cost_curve['delta_reward'].values / y_normalization_DRP, color='#FBB454', label='Covariate shift', linewidth=7)

# rDPR
plt.plot(x_index / x_normalization, rDRP_avg_aucc_cost_curve['delta_reward'].values / y_normalization, color='#40a368', label='No covariate shift', linewidth=7)


#plt.title('Insufficient data and No covariate shift')


plt.xlabel('Incremental cost', fontsize=12, fontweight='bold')
plt.ylabel('Incremental reward', fontsize=12, fontweight='bold')

ax = plt.gca()  # 

# 
ax.set_xticklabels([])
ax.set_yticklabels([])

# plt.xticks(fontsize=10)
# plt.yticks(fontsize=10)
plt.legend(prop={'size': 12, 'weight': 'bold'})
#plt.legend()
#plt.grid(True)
#plt.tight_layout()


# Save and display the figure

plt.savefig('../figure/limit_shift_.pdf', format='pdf', bbox_inches = 'tight')
plt.show()
```
